```python
import numpy as np
import pandas as pd
import random as rd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv(r"C:\Users\mehta\ML Codes\Lab 3\ex2data1-logistic.csv")
data1 = pd.read_csv(r"C:\Users\mehta\ML Codes\Lab 3\ex2data2-logistic.csv")
x1 = np.array(data['x1'])
x2 = np.array(data['x2'])
y = np.array(data['y'])
x1 = x1.reshape(100,1)
x2 = x2.reshape(100,1)
y = y.reshape(100,1)
x_1 = np.array(data1['x1'])
x_2 = np.array(data1['x2'])
y_y = np.array(data1['y'])
x_1 = x_1.reshape(118,1)
x_2 = x_2.reshape(118,1)
y_y = y_y.reshape(118,1)
x1 = (x1-np.mean(x1))/np.std(x1)
x2 = (x2-np.mean(x2))/np.std(x2)
x_1 = (x_1-np.mean(x1))/np.std(x_1)
x_2 = (x_2-np.mean(x2))/np.std(x_2)
```

```python
alpha = 0.001
theta = np.zeros([100000,3])
theta[0] = np.array([[0,20,20]])
import math
```

```python
for i in range(100000-1):
    index = rd.randint(0,89)
    x = np.array([1,x1[index],x2[index]])
    x = x.transpose()
    s = theta[i][0]*x[0] + theta[i][1]*x[1] + theta[i][2]*x[2]
    h = 1 / (1 + math.exp(-s))
    theta[i+1][0] = theta[i][0] - alpha*(h-y[index])
    theta[i+1][1] = theta[i][1] - (alpha*(h-y[index])*x1[index])
    theta[i+1][2] = theta[i][2] - (alpha*(h-y[index])*x2[index])
    

print(theta[i+1][0],theta[i+1][1],theta[i+1][2])
theta_0 = theta[i+1][0]
theta_1 = theta[i+1][1]
theta_2 = theta[i+1][2]
```

```python
plt.plot(x1,-(theta[i+1][0] + theta[i+1][1]*x1)/theta[i+1][2])
for i in range(len(x1)):
    if y[i]==1:
        plt.plot(x1[i],x2[i],'r*')
    else:
        plt.plot(x1[i],x2[i],'b*')
plt.title('Classification of Data Set 1')
plt.xlabel('x1')
plt.ylabel('x2')

plt.show()
```

```python
alpha = 0.0001
theta = np.zeros([1000000,6])
theta[0] = np.array([[0.5,0.5,0.5,0.5,0.5,0.5]])
for i in range(1000000-1):
    index = rd.randint(0,100)
    x = np.array([1,x_1[index]*x_1[index],x_2[index]*x_2[index],x_1[index],x_2[index],x_1[index]*x_2[index]])
    x = x.transpose()
    s = theta[i][0]*x[0] + theta[i][1]*x[1] + theta[i][2]*x[2] + theta[i][3]*x[3] + theta[i][4]*x[4] + theta[i][5]*x[5]
    h = 1 / (1 + math.exp(-s))
    theta[i+1][0] = theta[i][0] - alpha*(h-y_y[index])
    theta[i+1][1] = theta[i][1] - (alpha*(h-y_y[index])*x_1[index]*x_1[index])
    theta[i+1][2] = theta[i][2] - (alpha*(h-y_y[index])*x_2[index]*x_2[index])
    theta[i+1][3] = theta[i][3] - (alpha*(h-y_y[index])*x_1[index])
    theta[i+1][4] = theta[i][4] - (alpha*(h-y_y[index])*x_2[index])
    theta[i+1][5] = theta[i][5] - (alpha*(h-y_y[index])*x_2[index]*x_1[index])

print(theta[i+1])
theta_0 = theta[i+1][0]
theta_1 = theta[i+1][1]
theta_2 = theta[i+1][2]
theta_3 = theta[i+1][3]
theta_4 = theta[i+1][4]
theta_5 = theta[i+1][5]
```
